```python
#Predictive analysis
```

```python
dummies = pd.get_dummies(df['Company'])
```

```python
df = pd.concat([df, dummies], axis=1)
```

```python
df
```

```python
#drop charges
X=df.drop(columns=['MRP','Company','OpSys','Memory','No_of_ratings','Review'])
```

```python
X
```

```python
Y=df['MRP']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
mymodel=LinearRegression()
```

```python
df
```

```python
mymodel.fit(X_train,Y_train)
```

```python
mymodel.score(X_test,Y_test)
```

```python
mymodel.coef_
```

```python
mymodel.intercept_
```

```python
import pickle
```

```python
pickle.dump(mymodel,open("lineartest.pkl",'wb'))
```
